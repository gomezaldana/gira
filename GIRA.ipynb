#LENGUAJE GIRA

El objetivo del lenguaje es controlar dispositivos con velocidades y estados, podria usarse en distintos dispositivos como un centrifugador, una licuadora, un ventilador, entre otros.

El alcance del lenguaje abarca:

• Encendido y apagado del sistema.

• Configuración de velocidad dentro de rangos permitidos.

• Ejecución temporizada de procesos.

• Pausa e interrupción de la ejecución actual.

• Repetición de bloques de instrucciones.

• Uso de estructuras condicionales (si / sino).

• Uso de estructuras iterativas (mientras).

• Manejo de variables y valores simples.

• Evaluación de condiciones lógicas y comparaciones.

##ESPECIFICACIONES SINTACTICAS

PROGRAMA -> encender INSTRUCCIONES apagar

INSTRUCCIONES -> INSTRUCCIONES INSTRUCCION | INSTRUCCION

INSTRUCCION -> VELOCIDAD | EJECUTAR | PAUSAR | REPETIR | GIRAR | SI | MIENTRAS | id = VALOR

VELOCIDAD -> velocidad ( numero )

EJECUTAR -> ejecutar ( tiempo ) | ejecutar ( id )

PAUSAR -> pausar ( tiempo )

GIRAR -> girar ( numero , tiempo )

VALOR -> numero | tiempo | id | verdadero | falso

REPETIR -> repetir ( numero ) { INSTRUCCIONES }

SI  ->  si ( CONDICION ) { INSTRUCCIONES } | si ( CONDICION ) { INSTRUCCIONES } sino { INSTRUCCIONES }

MIENTRAS  ->  mientras ( CONDICION ) { INSTRUCCIONES }

CONDICION  ->  id | id > numero | id < numero | id == numero | verdadero | falso

##ESPECIFICACIONES SEMANTICAS

- El dispositivo debe estar encendido antes de usar

- El rango permitido de velocidad será de 0 a 10

- Los identificadores siempre empiezan con una letra

- Todo tiempo debe ser mayor a cero.

- repetir(numero) solo admite enteros positivos.

- Las condiciones deben comparar valores compatibles(temp > 20 se puede pero temp > encender no se puede)

- pausar() interrumpe la ejecución actual sin apagar el dispositivo.

- ejecutar(tiempo) daria un funcionamiento temporizado

- velocidad(numero) tiene un funcionamiento continuo si no pone ejecutar(tiempo)

##SCANNER

```python
%%writefile scanner_dispositivo.py

import ply.lex as lex

# ---------------- RESERVADAS ----------------

reservadas = {
    'encender': 'ENCENDER',
    'apagar': 'APAGAR',
    'velocidad': 'VELOCIDAD',
    'ejecutar': 'EJECUTAR',
    'pausar': 'PAUSAR',
    'repetir': 'REPETIR',
    'si': 'SI',
    'sino': 'SINO',
    'mientras': 'MIENTRAS',
    'verdadero': 'VERDADERO',
    'falso': 'FALSO',
    'girar': 'GIRAR',
}

# ---------------- TOKENS ----------------

tokens = [

    'ID',
    'NUMERO',
    'TIEMPO',

    'PARIZQ',
    'PARDER',
    'LLAVEIZQ',
    'LLAVEDER',
    'COMA',

    'MAS',
    'MENOS',
    'POR',
    'DIVIDIDO',

    'MAYOR',
    'MENOR',
    'MAYORIGUAL',
    'MENORIGUAL',
    'IGUALIGUAL',
    'DISTINTO',
    'IGUAL'

] + list(reservadas.values())

# ---------------- REGLAS ----------------

t_PARIZQ = r'\('
t_PARDER = r'\)'
t_LLAVEIZQ = r'\{'
t_LLAVEDER = r'\}'
t_COMA = r','

t_MAS = r'\+'
t_MENOS = r'-'
t_POR = r'\*'
t_DIVIDIDO = r'/'

t_MAYORIGUAL = r'>='
t_MENORIGUAL = r'<='
t_IGUALIGUAL = r'=='
t_DISTINTO = r'!='
t_MAYOR = r'>'
t_MENOR = r'<'
t_IGUAL = r'='

# tiempo
def t_TIEMPO(t):
    r'\d+(s|m|h)'
    return t

# numeros
def t_NUMERO(t):
    r'\d+'
    t.value = int(t.value)
    return t

# identificadores / reservadas
def t_ID(t):
    r'[a-zA-Z][a-zA-Z0-9]*'
    t.type = reservadas.get(t.value, 'ID')
    return t

t_ignore = ' \t'

# saltos
def t_newline(t):
    r'\n+'
    t.lexer.lineno += len(t.value)

# errores
def t_error(t):
    print("Carácter ilegal:", t.value[0])
    t.lexer.skip(1)

# construir lexer
lexer = lex.lex()
```

```python
%%writefile parser_dispositivo.py

import ply.yacc as yacc
from scanner_dispositivo import tokens
from scanner_dispositivo import lexer

hay_error = False

tabla_simbolos = {}

# ---------------- PROGRAMA ----------------

def p_programa(p):
    'programa : ENCENDER instrucciones APAGAR'
    global hay_error
    p[0] = not hay_error


def p_instrucciones(p):
    '''instrucciones : instrucciones instruccion
                     | instruccion'''


# ---------------- INSTRUCCIONES ----------------

def p_instruccion(p):
    '''instruccion : velocidad
                   | ejecutar
                   | pausar
                   | repetir
                   | girar
                   | si
                   | mientras
                   | asignacion'''


def p_velocidad(p):
    'velocidad : VELOCIDAD PARIZQ NUMERO PARDER'

    global hay_error, ultima_velocidad

    if p[3] < 0 or p[3] > 10:
        print("Error semántico: velocidad fuera de rango")
        hay_error = True

    ultima_velocidad = p[3]


def p_ejecutar(p):
    '''ejecutar : EJECUTAR PARIZQ TIEMPO PARDER
                | EJECUTAR PARIZQ ID PARDER'''

    global hay_error

    if p.slice[3].type == "TIEMPO":

        if str(p[3]).startswith("0"):
            print("Error semántico: tiempo inválido")
            hay_error = True

    else:

      if p[3] not in tabla_simbolos:
          print("Error semántico: variable no declarada")
          hay_error = True
          return

      valor = tabla_simbolos[p[3]]

      if not isinstance(valor, str) or not valor.endswith(("s", "m", "h")):
          print("Error semántico: la variable no contiene un tiempo")
          hay_error = True


def p_pausar(p):
    'pausar : PAUSAR PARIZQ TIEMPO PARDER'

def p_asignacion(p):
    'asignacion : ID IGUAL valor'

    tabla_simbolos[p[1]] = p[3]


def p_girar(p):
    'girar : GIRAR PARIZQ NUMERO COMA TIEMPO PARDER'


def p_valor(p):
    '''valor : NUMERO
             | TIEMPO
             | VERDADERO
             | FALSO'''

    p[0] = p[1]


def p_repetir(p):
    '''repetir : REPETIR PARIZQ NUMERO PARDER LLAVEIZQ instrucciones LLAVEDER'''

    global hay_error

    if isinstance(p[3], int) and p[3] <= 0:
        print("Error semantico: el numero para repetir debe ser positivo")
        hay_error = True

def p_si(p):
    '''si : SI PARIZQ condicion PARDER LLAVEIZQ instrucciones LLAVEDER
          | SI PARIZQ condicion PARDER LLAVEIZQ instrucciones LLAVEDER SINO LLAVEIZQ instrucciones LLAVEDER'''

def p_mientras(p):
    'mientras : MIENTRAS PARIZQ condicion PARDER LLAVEIZQ instrucciones LLAVEDER'


def p_condicion(p):
    '''condicion : ID
                 | ID MAYOR NUMERO
                 | ID MENOR NUMERO
                 | ID IGUALIGUAL NUMERO
                 | VERDADERO
                 | FALSO'''


# ---------------- ERROR ----------------

def p_error(p):
    global hay_error
    hay_error = True
    if p:
        print("Error sintáctico en:", p.value)
    else:
        print("Error sintáctico al final del archivo")


# ---------------- PARSER ----------------
parser = yacc.yacc()
```

```python
%%writefile test_parser.py
import unittest
import parser_dispositivo as pd
from scanner_dispositivo import lexer

class TestParser(unittest.TestCase):

    def probar_codigo(self, texto):
        pd.hay_error = False
        resultado = pd.parser.parse(texto, lexer=lexer)
        if resultado is None:
            return False
        return resultado

    # ---------------- VALIDOS ----------------



    def test_completo(self):
        codigo = """
        encender

        modo = verdadero

        ciclos = 3

        tiempoLento = 10s

        tiempoRapido = 5s

        temperatura = 25

        si(modo){

        repetir(3){

        velocidad(2)

        ejecutar(tiempoLento)

        pausar(2s)

        velocidad(5)

        ejecutar(tiempoRapido)

        girar(180, 3s)

        }

        }

        sino{

          velocidad(1)

          ejecutar(20s)

        }

        mientras(temperatura > 20){

          velocidad(3)

          ejecutar(5s)

          pausar(1s)

        }

        apagar
        """
        self.assertTrue(self.probar_codigo(codigo))


    def test_repetir(self):
        codigo = """
        encender

        tiempo = 10m

        esCentrifugado = falso

        si(esCentrifugado){

          repetir(5){

          velocidad(2)

          ejecutar(tiempo)

          velocidad(4)

          ejecutar(tiempo)
          }

        }

        sino{

          velocidad(6)

          ejecutar(5m)
        }

        apagar
        """
        self.assertTrue(self.probar_codigo(codigo))

    # ---------------- INVALIDOS ----------------

    def test_tiempo_mal(self):
        codigo = """
        encender
        uno=1
        repetir(2){

            velocidad(1)

            ejecutar(uno)

            pausar(30s)

            velocidad(3)

            ejecutar(4m)

            pausar(30s)

            velocidad(5)

            ejecutar(4m)

        }

        apagar


        """
        self.assertFalse(self.probar_codigo(codigo))


    def test_falta_apagar(self):
        codigo = """
        encender
        velocidad(3)
        """
        self.assertFalse(self.probar_codigo(codigo))

    def test_asignacion_falla_velocidad(self):
        codigo = """
        encender
          veloc = 135
          repetir(3){
            velocidad(veloc)
            ejecutar(13m)
            si(verdadero){
              pausar(2s)
              veloc = 2
              }
            }
          }

        apagar
        """
        self.assertFalse(self.probar_codigo(codigo))

    def test_parentesis_mal(self):
        codigo = """
        encender
        velocidad(3
        apagar
        """
        self.assertFalse(self.probar_codigo(codigo))

    def test_repetir_mal(self):
        codigo = """
        encender
        repetir(x){
            velocidad(2)
        }
        apagar
        """
        self.assertFalse(self.probar_codigo(codigo))


if __name__ == '__main__':
    unittest.main(verbosity=2)
```
